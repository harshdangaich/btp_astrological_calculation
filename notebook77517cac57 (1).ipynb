```python
import pandas as pd
from kerykeion import AstrologicalSubject, Report

# Read the CSV file into a pandas DataFrame
df = pd.read_csv("/kaggle/input/xxxxxx/btp_data _final.txt")

# Iterate through each row of the DataFrame
for _, row in df.iterrows():
    # Create an AstrologicalSubject instance for each person
    astro_subject = AstrologicalSubject(row["Name"], row["Year"], row["Month"], row["Day"], row["Hour"], row["Minutes"], row["City"])

    # Generate an astrological report for the person
    report = Report(astro_subject)

    # Print the report in the desired format
    print(f"+- Kerykeion report for {row['Name']} -+")
    report.print_report()
    print("\n")
```
